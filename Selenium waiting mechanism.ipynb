1.1 什么是隐式等待？
隐式等待是一种全局性的等待机制，它会在查找元素时等待一定的时间。如果在指定的时间内找到了元素，Selenium 会立即继续执行后续操作；如果超时仍未找到元素，则会抛出 NoSuchElementException 异常。

1.2 如何使用隐式等待？
隐式等待通过 implicitly_wait() 方法来设置。这个方法只需要调用一次，之后的所有元素查找操作都会遵循这个等待时间。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
import time
# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

driver.implicitly_wait(10)     #隐式等待不会主动等待10秒，它只是在查找元素时如果没有立即找到，会等待最多10秒。


try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待搜索按钮可点击
    
    button_element = driver.find_element(By.ID, "fill-btn")
    button_element.click()
    print("✓ 已点击搜索按钮")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

2.1 什么是显式等待？
显式等待是一种更为灵活的等待机制，它允许你为特定的操作设置等待条件。显式等待通常与 WebDriverWait 类和 expected_conditions 模块一起使用。

2.2 如何使用显式等待？
显式等待的基本用法如下：

常用的 expected_conditions
expected_conditions 模块提供了多种预定义的等待条件，以下是一些常用的条件：

presence_of_element_located：等待元素出现在 DOM 中。
visibility_of_element_located：等待元素出现在 DOM 中并且可见。
element_to_be_clickable：等待元素可点击。
text_to_be_present_in_element：等待元素的文本包含指定的文本。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
import time
# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待搜索按钮可点击
    wait = WebDriverWait(driver, 10)
    button_element =  wait.until(EC.presence_of_element_located((By.ID, "fill-btn")))
    button_element.click()
    print("✓ 已点击搜索按钮")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

3.1 什么是固定等待？
固定等待是一种最简单的等待机制，它通过 time.sleep() 方法让脚本暂停执行指定的时间。无论页面是否加载完成，脚本都会等待指定的时间后再继续执行。

3.2 如何使用固定等待？

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
import time
# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    start_time = time.time()
    time.sleep(10)
    # 等待搜索按钮可点击
    button_element = driver.find_element(By.ID, "fill-btn")
    button_element.click()
    end_time = time.time()
    print(f"花费时间:{end_time - start_time:.2f} 秒")
    print("✓ 已点击搜索按钮")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

Fluent Wait
作用：动态设置等待条件，可以自定义轮询频率和忽略特定异常。

方法：WebDriverWait 结合 polling_every 和 ignoring

特点：更灵活，适用于需要动态调整等待时间的场景。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.common.exceptions import NoSuchElementException
import time
# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    start_time = time.time()
    # 等待搜索按钮可点击
    wait = WebDriverWait(driver, timeout=10, poll_frequency=1, ignored_exceptions=[NoSuchElementException])
    button_element = wait.until(EC.presence_of_element_located((By.ID, "fill-btn")))
    button_element.click()
    end_time = time.time()
    print(f"花费时间:{end_time - start_time:.2f} 秒")
    print("✓ 已点击搜索按钮")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```
