```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('https://www.baidu.com')
    print("✓ 已访问百度首页")
    
    # 方法1：等待元素可点击（更严格）
    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.presence_of_element_located((By.ID, "chat-textarea")))
    print("✓ 搜索框已可交互")
    
    # 输入内容
    element.send_keys("Selenium 教程")
    print("✓ 已输入搜索内容")
    
    # 等待搜索按钮可点击
    search_button = wait.until(EC.element_to_be_clickable((By.ID, "chat-submit-button")))
    wait = WebDriverWait(driver, 10)
    search_button.click()
    print("✓ 已点击搜索按钮")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 方法1：等待元素可点击（更严格）
    wait = WebDriverWait(driver, 10)
    element = wait.until(EC.presence_of_element_located((By.NAME, "phone-number")))
    print("✓ 搜索框已可交互")

    # 获取各种属性值
    class_value = element.get_attribute("class")
    id_value = element.get_attribute("id")
    name_value = element.get_attribute("name")
    type_value = element.get_attribute("type")
    placeholder_value = element.get_attribute("placeholder")
    value_value = element.get_attribute("value")
    
    # 输入内容
    print("✓ 已获取元素的属性值")

    # 打印类型和值
    print(f"\n--- 属性值类型 ---")
    print(f"class 类型: {type(class_value)}, 值: {class_value}")
    print(f"id 类型: {type(id_value)}, 值: {id_value}")
    print(f"name 类型: {type(name_value)}, 值: {name_value}")
    print(f"type 类型: {type(type_value)}, 值: {type_value}")
    print(f"placeholder 类型: {type(placeholder_value)}, 值: {placeholder_value}")
    print(f"value 类型: {type(value_value)}, 值: {value_value}")
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

```python

from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 方法1：等待元素可点击（更严格）
    wait = WebDriverWait(driver, 10)
    checkbox_element = wait.until(EC.presence_of_element_located((By.ID, "check1")))
    print("✓ 复选框已可交互")
    
    # 如果复选框未被选中，则点击选中
    if not checkbox_element.is_selected():
        checkbox_element.click()
    print(f"✓ 复选框是否选中：{checkbox_element.is_selected()}")
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问百度
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 方法1：等待元素可点击（更严格）
    wait = WebDriverWait(driver, 10)
    dropdown_element = wait.until(EC.presence_of_element_located((By.ID, "dynamic-select")))
    print("✓ 下拉列表已可交互")
    
    #打印原始索引值
    print(f"原始索引值：{dropdown_element.is_selected()}")

    # 创建Select对象
    select = Select(dropdown_element)
    
    # 通过索引选择选项
    select.select_by_index(2)

    #打印更改后索引值
    print(f"更改后索引值：{dropdown_element.is_selected()}")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```
