Selenium 实战项目

在现代软件开发中，自动化测试已经成为确保软件质量的重要手段之一。

Selenium 是一个广泛使用的自动化测试工具，它支持多种编程语言和浏览器，能够模拟用户操作，进行功能测试、回归测试等。

本文将详细介绍如何使用 Selenium 进行自动化测试，并通过实战项目演示如何实现登录、注册和搜索（以百度为例）等常见功能的自动化测试。

1. 实战项目：自动化测试案例

1.1 登录功能测试

假设我们有一个登录页面，包含用户名和密码输入框以及登录按钮，我们将编写一个自动化测试脚本来模拟用户登录操作。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.common.exceptions import TimeoutException
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_argument("--headless")  # 启用无头模式
edge_options.add_argument("--disable-gpu")  # 禁用 GPU 加速
edge_options.add_argument("--window-size=1920,1080")  # 设置窗口大小
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 打开登陆界面
    driver.get('http://localhost:8080/login.html')
    print("✓ 已访问登录页面")
    
    # 等待页面加载
    wait = WebDriverWait(driver, 10)
    
    # 定位用户名输入框并输入用户名
    username_input = wait.until(EC.presence_of_element_located((By.ID, "username")))
    print("✓ 已定位用户名输入框")
    username_input.send_keys("test")
    
    # 定位密码输入框并输入密码
    password_input = driver.find_element(By.ID, "password")
    print("✓ 已定位密码输入框")
    password_input.send_keys("991227")
    
    # 定位登录按钮并点击
    button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@type='submit' and text()='登录']")))
    print("✓ 已定位登录按钮")
    button.click()
    print("✓ 已点击登录按钮")
    
    # 等待登录后的页面加载
    time.sleep(2)  # 等待页面跳转
    
    # 验证登录成功（等待删除按钮出现）
    try:
        delete_button = wait.until(EC.presence_of_element_located((By.ID, "deleteBtn")))
        assert delete_button.is_displayed(), "删除按钮不可见，登录可能失败"
        print("✓ 登录成功，找到删除按钮")
    except TimeoutException:
        print("✗ 登录失败：未找到删除按钮")
        # 打印当前页面信息用于调试
        print(f"当前 URL: {driver.current_url}")
        print(f"页面标题: {driver.title}")
        raise
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

1.2 注册功能测试

接下来，我们编写一个自动化测试脚本来模拟用户注册操作。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.common.exceptions import TimeoutException
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_argument("--headless")  # 启用无头模式
edge_options.add_argument("--disable-gpu")  # 禁用 GPU 加速
edge_options.add_argument("--window-size=1920,1080")  # 设置窗口大小
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 打开登陆界面
    driver.get('http://localhost:8080/register.html')
    print("✓ 已访问注册页面")
    
    # 等待页面加载
    wait = WebDriverWait(driver, 10)
    
    # 定位用户名输入框并输入用户名
    username_input = driver.find_element(By.ID, "username")
    print("✓ 已定位用户名输入框")
    username_input.send_keys("test111")

    # 定位邮箱输入框并输入邮箱
    email_input = driver.find_element(By.ID, "email")
    print("✓ 已定位邮箱输入框")
    email_input.send_keys("test@example.com")
    
    # 定位密码输入框并输入密码
    password_input = driver.find_element(By.ID, "password")
    print("✓ 已定位密码输入框")
    password_input.send_keys("991227")


    # 再次定位密码输入框并输入密码
    password_input = driver.find_element(By.ID, "confirm_password")
    print("✓ 已再次定位密码输入框")
    password_input.send_keys("991227")
    
    # 定位注册按钮并点击
    button = driver.find_element(By.XPATH, "//button[@type='submit' and text()='注册']")
    print("✓ 已定位注册按钮")
    button.click()
    print("✓ 已点击注册按钮")
    
    # 等待注册后的页面加载
    time.sleep(2)  # 等待页面跳转
    
    # 验证登录成功（等待删除按钮出现）
    try:
        delete_button = driver.find_element(By.XPATH, "//button[@type='submit' and text()='登录']")
        assert delete_button.is_displayed(), "登录按钮不可见，注册可能失败"
        print("✓ 注册成功，找到登录按钮")
    except TimeoutException:
        print("✗ 登录失败：未找到删除按钮")
        # 打印当前页面信息用于调试
        print(f"当前 URL: {driver.current_url}")
        print(f"页面标题: {driver.title}")
        raise
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.keys import Keys
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_argument("--headless")  # 启用无头模式
edge_options.add_argument("--disable-gpu")  # 禁用 GPU 加速
edge_options.add_argument("--window-size=1920,1080")  # 设置窗口大小
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 打开登陆界面
    driver.get('http://baidu.com')
    print("✓ 已访问百度页面")
    
    search_box = driver.find_element(By.ID, "chat-textarea")
    print("✓ 已找到搜索框")
    search_box.send_keys("runoob")
    print("✓ 已输入搜索内容")

    #按下回车
    search_box.send_keys(Keys.RETURN)
    print("✓ 已按下搜索按钮")
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```
