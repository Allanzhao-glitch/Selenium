Selenium 鼠标和键盘操作
Selenium 提供了丰富的鼠标和键盘操作功能，可以模拟用户的交互行为，例如点击、双击、拖放、悬停、输入文本、按下组合键等。


鼠标操作（ActionChains 类）
在 Selenium 中，鼠标操作主要通过 ActionChains 类来实现。ActionChains 类允许我们执行复杂的鼠标操作，如点击、双击、右键点击、拖放和悬停等。

点击、双击、右键点击

点击（Click）：模拟鼠标左键点击操作。

双击（Double Click）：模拟鼠标左键双击操作。

右键点击（Right Click）：模拟鼠标右键点击操作。


```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.action_chains import ActionChains

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

driver.implicitly_wait(10)     #隐式等待不会主动等待10秒，它只是在查找元素时如果没有立即找到，会等待最多10秒。


try:
    # 访问首页
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待搜索按钮可点击
    
    button_element = driver.find_element(By.ID, "fill-btn")
    
    print("✓ 已找到搜索按钮")
    #创建ActionChains对象
    actions = ActionChains(driver)
    #点击操作
    actions.click(button_element).perform()
    print("✓ 点击操作")
    
    # 双击操作
    actions.double_click(button_element).perform()
    print("✓ 双击操作")
    # 右键点击操作
    actions.context_click(button_element).perform()
    print("✓ 右键点击操作")

    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

拖放操作

拖放操作是指将一个元素拖动到另一个位置或另一个元素上。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.action_chains import ActionChains

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

driver.implicitly_wait(10)     #隐式等待不会主动等待10秒，它只是在查找元素时如果没有立即找到，会等待最多10秒。


try:
    # 访问首页
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待搜索按钮可点击
    
    drag1 = driver.find_element(By.ID, "drag1")
    drop_zone = driver.find_element(By.ID,"drop-result")

    # 创建 ActionChains 对象
    actions = ActionChains(driver)

    # 拖放操作
    actions.drag_and_drop(drag1, drop_zone).perform()

    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

悬停操作

悬停操作是指将鼠标指针悬停在某个元素上，通常用于触发下拉菜单或显示提示信息。

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By

# 初始化浏览器驱动
driver = webdriver.Chrome()

# 打开网页
driver.get("https://example.com")

# 定位元素
element = driver.find_element(By.ID, "element_id")

# 创建 ActionChains 对象
actions = ActionChains(driver)

# 悬停操作
actions.move_to_element(element).perform()

# 关闭浏览器
driver.quit()
```

键盘操作（Keys 类）

Selenium 中的键盘操作主要通过 Keys 类来实现。Keys 类提供了各种键盘按键的模拟操作，包括发送组合键和特殊键操作。

发送组合键

组合键操作是指同时按下多个键，例如 Ctrl + C（复制）或 Ctrl + V（粘贴）。

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 设置正确的驱动路径
service = ChromeService(executable_path="./chromedriver-mac-arm64/chromedriver")
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# 打开网页
driver.get("https://example.com")

# 定位输入框
input_element = driver.find_element(By.ID, "input_element_id")

# 输入内容并发送组合键
input_element.send_keys("Hello, World!")
input_element.send_keys(Keys.CONTROL, 'a')  # 全选
input_element.send_keys(Keys.CONTROL, 'c')  # 复制
input_element.send_keys(Keys.CONTROL, 'v')  # 粘贴

# 关闭浏览器
driver.quit()
```

特殊键操作（Enter、Tab、Shift 等）

特殊键操作是指模拟按下键盘上的特殊键，如 Enter、Tab、Shift 等。

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# 设置正确的驱动路径
service = ChromeService(executable_path="./chromedriver-mac-arm64/chromedriver")
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# 打开网页
driver.get("https://example.com")

# 定位输入框
input_element = driver.find_element(By.ID, "input_element_id")

# 输入内容并按下 Enter 键
input_element.send_keys("Hello, World!")
input_element.send_keys(Keys.ENTER)

# 按下 Tab 键切换到下一个元素
input_element.send_keys(Keys.TAB)

# 按下 Shift 键并输入大写字母
input_element.send_keys(Keys.SHIFT, "a")  # 输入大写的 'A'

# 关闭浏览器
driver.quit()
```
