1. 处理浏览器弹窗（alert、confirm、prompt）
在 Web 应用程序中，弹窗（如 alert、confirm、prompt）是常见的用户交互方式。Selenium 提供了 Alert 类来处理这些弹窗。

1.1 处理 alert 弹窗
alert 弹窗通常用于显示一条消息，用户只能点击"确定"按钮关闭弹窗。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.alert import Alert
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问页面
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待页面完全加载
    time.sleep(2)
    
    # 触发 alert 弹窗
    driver.execute_script("alert('这是一个 alert 弹窗');")
    print("✓ 触发 alert 弹窗")
    
    # 等待弹窗出现
    time.sleep(1)
    
    # 切换到 alert 弹窗
    alert = Alert(driver)
    print("✓ 切换到 alert 弹窗")
    
    # 获取弹窗文本
    print(f"✓ 获取弹窗文本: {alert.text}")
    time.sleep(10)
    # 点击"确定"按钮关闭弹窗
    alert.accept()
    print("✓ 点击确定按钮关闭弹窗")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

1.3 处理 prompt 弹窗

prompt 弹窗通常用于获取用户输入，用户可以输入文本并选择"确定"或"取消"。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.alert import Alert
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问页面
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待页面完全加载
    time.sleep(2)
    
    # 触发 alert 弹窗
    driver.execute_script("prompt('这是一个 prompt 弹窗');")
    print("✓ 触发 prompt 弹窗")
    
    # 等待弹窗出现
    time.sleep(1)
    
    # 切换到 alert 弹窗
    alert = Alert(driver)
    print("✓ 切换到 prompt 弹窗")
    
    # 获取弹窗文本
    print(f"✓ 获取弹窗文本: {alert.text}")

    alert.send_keys("用户输入")
    print("✓ 输入用户输入")

    # 点击"确定"按钮关闭弹窗
    alert.accept()
    print("✓ 点击确定按钮关闭弹窗")
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

2. 切换窗口和标签页

在 Web 应用程序中，用户可能会打开多个窗口或标签页。Selenium 提供了 switch_to.window() 方法来切换窗口或标签页。

2.1 切换窗口

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.alert import Alert
import time

# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# 创建驱动
driver = webdriver.Edge(options=edge_options)

try:
    # 访问页面
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已访问首页")
    
    # 等待页面完全加载
    time.sleep(2)
    
    # 打开第二个窗口
    driver.execute_script("window.open('https://baidu.com');")

    # 获取所有窗口句柄
    window_handles = driver.window_handles

    # 切换到第二个窗口
    driver.switch_to.window(window_handles[1])

    # 在第二个窗口中操作
    print(driver.title)

    # 切换回第一个窗口
    driver.switch_to.window(window_handles[0])

    # 在第二个窗口中操作
    print(driver.title)
    
    # 保持浏览器打开
    input("按回车键关闭浏览器...")
    
except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

2.2 切换标签页

切换标签页与切换窗口类似，因为每个标签页都是一个独立的窗口句柄。

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.edge.options import Options
# 配置 Edge 选项
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

driver = webdriver.Edge(options=edge_options)

try:
    # 打开第一个标签页
    driver.get('http://localhost:8080/selenium-demo.html')
    print("✓ 已打开第一个标签页")
    
    # 打开第二个标签页
    driver.execute_script("window.open('https://www.baidu.com')")

    #获取所有标签页句柄
    tab_handles = driver.window_handles

    #切换到第二个标签页
    driver.switch_to.window(tab_handles[1])

    #在第二个标签页中操作
    print(f"✓ 已打开第二个标签页:{driver.title}")

    #切换到第1个标签页
    driver.switch_to.window(tab_handles[0])

    #在第1个标签页中操作
    print(f"✓ 已打开第一个标签页:{driver.title}")



except Exception as e:
    print(f"发生错误: {e}")
    
finally:
    driver.quit()
```

3. 处理 iframe

iframe 是嵌入在网页中的另一个网页。Selenium 提供了 switch_to.frame() 方法来切换到 iframe。

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService

# 设置正确的驱动路径
edge_options = Options()
edge_options.add_experimental_option("excludeSwitches", ["enable-logging"])

driver = webdriver.Edge(options=edge_options)

# 打开网页
driver.get("https://example.com")

# 切换到 iframe
iframe = driver.find_element(By.TAG_NAME, "iframe")
driver.switch_to.frame(iframe)

# 在 iframe 中操作
print(driver.page_source)

# 切换回主页面
driver.switch_to.default_content()

# 关闭浏览器
driver.quit()
```
